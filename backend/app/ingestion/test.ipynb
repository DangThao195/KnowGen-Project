```python
import os
import sys
import json
from IPython.display import display, Markdown

# 0. Sửa lỗi "ModuleNotFoundError: No module named 'app'" khi chạy trong thư mục con
sys.path.append(os.path.abspath("../../"))

from app.ingestion.extract import Document
from app.ingestion.transform import transform_documents

json_file = "extracted_metadata.json"

# 1. Nạp dữ liệu
try:
    with open(json_file, "r", encoding="utf-8") as f:
        raw_dicts = json.load(f)
    print(f"✅ Đã tải file JSON ({len(raw_dicts)} bản ghi)\n")
except FileNotFoundError:
    print(f"❌ Không tìm thấy file {json_file}.")
    raw_dicts = []

# ==========================================
# 2. CHUYỂN JSON DICT THÀNH OBJECT PYDANTIC
# ==========================================
documents_to_transform = [Document(**d) for d in raw_dicts]

# 3. Test hàm Transform
print("🔄 Đang chạy Pipeline Transform (Clean -> Summarize -> Chunk)...\n")
transformed_results = transform_documents(documents_to_transform)

# 4. Hiển thị kết quả Transform
print(f"✅ Đã xong! Transform được {len(transformed_results)} tài liệu.\n")

for i, t_doc in enumerate(transformed_results):
    print("="*80)
    print(f"🚀 KẾT QUẢ TRANSFORM TÀI LIỆU SỐ {i+1} : {t_doc.title}")
    print(f"📦 Tổng số chunks : {len(t_doc.chunks)}")
    print("-"*80)
    
    # In chi tiết từng chunk ra xem thử
    for chunk in t_doc.chunks:
        role = chunk.metadata.get('role', 'unknown')
        header_path = chunk.metadata.get('header_path', 'N/A')
        
        print(f"🔹 [Chunk {chunk.chunk_index}] - Role: {role.upper()} | Path: {header_path}")
        # Rút gọn nội dung (in 150 ký tự đầu tiên để test)
        preview = chunk.content[:300].replace('\n', ' ') + "..."
        print(f"   Nội dung: {preview}\n")
```
